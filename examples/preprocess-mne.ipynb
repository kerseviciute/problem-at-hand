```python
covariance = np.cov(data)
```

```python
covariance.shape
```

```python
eigenvalues, eigenvectors = np.linalg.eig(covariance)
```

```python
positions = finger.get_montage().get_positions()[ 'ch_pos' ]
positions = np.array(list(positions.values()))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

# Scatter plot
ax.scatter(positions[ :, 0 ], positions[ :, 1 ], positions[ :, 2 ], c = eigenvectors[ :, 0 ])

# Set labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
```
